## Storing Data
Save gathered, assessed, and cleaned master dataset to a CSV file named "twitter_archive_master.csv".

```python
twitter_archive_master = pd.merge(twitter_archive_enhanced_Original, image_predictions_Original, how = 'left', on = ['tweet_id'] )
#twitter_archive_master= pd.merge(twitter_archive_master, tweet_json_Original, how = 'left', on = ['id'])
twitter_archive_master.to_csv('twitter_archive_master.csv', encoding = 'utf-8')
twitter_archive_master.info()
```

```python
# Extract the dog stage names from the the 4 individual columns and combine into one 'Stage' column
stages = ['doggo', 'floofer', 'pupper', 'puppo']
for stage in stages:
    twitter_archive_master[stage] = twitter_archive_master[stage].apply(lambda x: np.NaN if x == 'None' else x)
    
twitter_archive_master['stage'] = twitter_archive_master[['doggo', 'floofer', 'pupper', 'puppo']].astype(str).sum(1)
twitter_archive_master['stage'] = twitter_archive_master['stage'].apply(lambda x: x.replace('nan', ''))
twitter_archive_master['stage'] = twitter_archive_master['stage'].apply(lambda x: np.NaN if x == '' else x)

twitter_archive_master['stage'].replace({'doggopupper':'doggo pupper', 
                                'doggofloofer':'doggo floofer',
                                'doggopuppo':'doggo puppo',
                                '':None}, inplace=True)
```

```python
# Drop individual dog stage columns
columns_to_drop = ['doggo', 'floofer', 'pupper', 'puppo']
twitter_archive_master.drop(columns_to_drop, axis=1, inplace=True)
```
